### Заполняем пропуски в координатах

```python
geolocator = Nominatim(user_agent="buildings_geocoder")

def geocode_row(row):
    if row['coordinates'] != 'не указано':
        return row['coordinates']
    
    address = f"{row['number']} {row['street']}, {row['city']}, Приморский край, Россия"
    
    try:
        location = geolocator.geocode(address)
        if location:
            coords_str = f"{{'lat':{location.latitude}, 'lon':{location.longitude}}}"
            return coords_str
        else:
            return 'не указано'
    except Exception:
        return 'не указано'

geocode_func = RateLimiter(geocode_row, min_delay_seconds=1)

mask_empty = buildings_cities_ref['coordinates'] == 'не указано'
buildings_cities_ref.loc[mask_empty, 'coordinates'] = buildings_cities_ref.loc[mask_empty].apply(geocode_func, axis=1)

print("Новые координаты:")
buildings_cities_ref[buildings_cities_ref['coordinates'] != 'не указано']
```

### Заполняем пропуски в big_folk_district_name, folk_district_name, district_name используя самые распространённые по отношению к street, так как в этой колонке нет NaN

```python
def fill_na(group):
    for col in ['big_folk_district_name', 'folk_district_name', 'district_name']:
        if group[col].notna().any():
            mode_values = group[col].mode()
            if len(mode_values) > 0:
                group[col] = group[col].fillna(mode_values[0])
    return group

buildings_cities_ref = buildings_cities_ref.groupby('street').apply(fill_na).reset_index(drop=True)
buildings_cities_ref
```

```python

def parse_coords(coord_str):
    if not coord_str or coord_str in ['[]', 'не указано', '']:
        return None, None
    try:
        fixed_str = coord_str.split(', c')[0].strip() if 'c' in coord_str.lower() else coord_str
        fixed_str = fixed_str.replace("'", '"')
        parsed = json.loads(fixed_str)
        if isinstance(parsed, list) and parsed:
            d = parsed[0]
            return float(d.get('lat')), float(d.get('lon'))
        elif isinstance(parsed, dict):
            return float(parsed.get('lat')), float(parsed.get('lon'))
        else:
            return None, None
    except Exception as e:
        print(f"Ошибка парсинга: {coord_str} - {e}")
        return None, None

buildings_cities_ref['lat'], buildings_cities_ref['lon'] = zip(*buildings_cities_ref['coordinates'].apply(parse_coords))

valid_df = buildings_cities_ref.dropna(subset=['lat', 'lon'])

if not valid_df.empty:
    m = folium.Map(location=[43.35, 132.18], zoom_start=10, tiles='OpenStreetMap')
    
    heat_data = [[row['lat'], row['lon']] for _, row in valid_df.iterrows()]
    
    HeatMap(heat_data, radius=15, blur=20, min_opacity=0.3).add_to(m)
    
    display(m)
else:
    print("Нет валидных координат для карты.")
```
